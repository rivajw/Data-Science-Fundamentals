b) What are some limitations of K means ++?

- Kmeans++ tend to favor cluster that have similar size
- If you have a high density region, Kmeans++ tends to clump the points in the region together instead of seperating them into clusters. However, everything is at random, so this doesn't happen "all the time".
- Does not do well with non-globular shape
